# LOAD/DISPLAY IMAGE

```python
import cv2
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
image_color = cv2.imread('watchingTV_moving_car.jpg')

# image_color = cv2.imread('red_night.jpg')
# image_color = cv2.imread('red_car_2.jpg')

cv2.imshow('Original Image', image_color)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
channels = cv2.split(image_color)
colors = ('b', 'g', 'r')
plt.figure()
plt.title("color histogram")
plt.xlabel("bins")
plt.ylabel("# of pixels")
features = []

for (channel, color) in zip(channels, colors):
    hist = cv2.calcHist([channel], [0], None, [256], [0,256])
    #features.extend(hist)
    plt.plot(hist, color=color)
    plt.xlim([0, 256])


plt.show()
```
